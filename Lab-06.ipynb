```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = {
    'Employee_ids': [101, 102, 103, 104, 105],
    'Genders': ['M', 'F', 'M', 'T', 'M'],
    'age': [24, 20, 40, 30, 27]
}
df = pd.DataFrame(data)
label_encoder = LabelEncoder()
df['Gender_Encoded'] = label_encoder.fit_transform(df['Genders'])

one_hot_encoded = pd.get_dummies(df['Genders'], prefix='Genders',dtype=int)
df = pd.concat([df, one_hot_encoded], axis=1)
display(df)
```

```python
from datetime import datetime

data = {'Name': ['sabir', 'sai','trishala'], 'join year': ['2024', '2023','2022']}
df = pd.DataFrame(data)

df['join year'] = pd.to_datetime(df['join year'])

current_year = datetime.now().year
df['current year'] = current_year - df['join year'].dt.year
display(df)
```

```python
data = {'Category': ['A', 'A', 'B', 'B', 'C'], 'Value': [10, 20, 10, 30, 40]}
df = pd.DataFrame(data)

grouped = df.groupby('Category').agg({'Value': ['sum', 'mean']})
display( grouped)

pivot = df.pivot_table(index='Category', values='Value', aggfunc=['sum', 'mean'])
display( pivot)
```
