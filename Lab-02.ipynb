```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Histogram of Age and passengerId

```python
plt.hist(df['Age'],bins=9,linewidth=0.5, edgecolor="white", color='green')
plt.xlabel('Age')
plt.ylabel('PassengerId')
plt.title('Histogram of Age')
plt.show()
```

Bar chart on Age and Embarked

```python
plt.bar(df['Embarked'].astype(str) ,df['Age'],color='skyblue',)
plt.xlabel('Embarked')
plt.ylabel('Age')
plt.title('Bar Chart of Age')
plt.show()
```

```python
sns.boxplot(df['Age'] ,color='pink')
plt.xlabel('Age')
plt.title('Box Plot of Age')
plt.ylim(-10,100)
plt.show()
```

```python
plt.figure(figsize=(80, 5))
sns.boxplot(x=df['Cabin'], y=df['Age'], color='blue')
plt.xlabel('Cabin')
plt.ylabel('Age')
plt.title('Box Plot of Age by Cabin')
plt.ylim(-10,100)
plt.show()
```
